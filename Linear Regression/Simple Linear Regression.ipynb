### Handling interactive datas

```python
# Load libraries 
from sklearn.linear_model import LinearRegression 
from sklearn.datasets import load_boston
from sklearn.preprocessing import PolynomialFeatures
# Load data with only two features 
boston = load_boston() 
features = boston.data[:,0:2] 
target = boston.target
```

```python
df_hid = pd.DataFrame(features)
df_hid.columns = boston.feature_names[0:2]
df_hid.head(2)
```

```python
#create interacive term 
interaction = PolynomialFeatures(degree=3, include_bias=False, 
                                 interaction_only=True) 
features_interaction = interaction.fit_transform(features)
```

```python
df_intcr = pd.DataFrame(features_interaction)
df_intcr.head()
```

```python
#creating polynimial feature x^2 and x^3
polynomial = PolynomialFeatures(degree=3, include_bias=False) 
features_polynomial = polynomial.fit_transform(features)
```

```python
df_intcr1 = pd.DataFrame(features_polynomial)
df_intcr1.head()
```
