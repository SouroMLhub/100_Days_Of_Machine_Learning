```python
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
oli = datasets.fetch_olivetti_faces()
```

```python
oli.keys()
```

```python
oli['data'].shape
```

```python
oli['images'].shape
```

```python
fig = plt.figure(figsize=(8,8))
for i in range(64):#8 rows and 8 column 
    ax = fig.add_subplot(8, 8, i+1)#row, coloum and position
    ax.imshow(oli.images[i], cmap = plt.cm.bone)
plt.show()
```

```python
x = oli.data
y = oli.target
```

```python
#applying pca
pca = PCA()
pca.fit(x)
```

```python
pca.components_.shape
```

```python
pca.explained_variance_
```

```python
#figuring out number of component
k = 0
total = sum(pca.explained_variance_)
currentSum = 0
while currentSum/total<0.95:
    currentSum +=pca.explained_variance_[k]
    k=k+1
k
```

```python
pca = PCA(n_components=k)
transformed_data = pca.fit_transform(x)
transformed_data.shape
```

```python
x_approx = pca.inverse_transform(transformed_data)
x_approx.shape
```

```python
x_approx_images = x_approx.reshape((400, 64, 64))
```

```python
fig = plt.figure(figsize=(8,8))
for i in range(64):#8 rows and 8 column 
    ax = fig.add_subplot(8, 8, i+1)#row, coloum and position
    ax.imshow(x_approx_images[i], cmap = plt.cm.bone)
plt.show()
```

```python
eignvectors = pca.components_
eignvectors.shape
```

```python
eigenfaces = eignvectors.reshape((123, 64, 64))
```

```python
fig = plt.figure(figsize=(8,8))
for i in range(64):  
    ax = fig.add_subplot(8, 8, i + 1)
    ax.imshow(eigenfaces[i], cmap=plt.cm.bone)
plt.show()
```
