```python
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import ensemble
import time
```

```python
lfw = datasets.fetch_lfw_people(min_faces_per_person=100, resize = 0.4)
```

```python
lfw.keys()
```

```python
lfw['data'].shape
```

```python
lfw['images'].shape
```

```python
fig = plt.figure(figsize = (8,8))
for i in range(64):
    ax = fig.add_subplot(8, 8, i+1)
    ax.imshow(lfw.images[i], cmap = plt.cm.bone)
plt.show()
```

```python
X = lfw.data
Y = lfw.target
```

```python
pca = PCA()
pca.fit(X)
```

```python
pca.components_
```

```python
x, y = lfw.data, lfw.target
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state =0)
pca_oliv = PCA()
pca_oliv.fit(x_train)
```

```python
k = 0
total = 0
while total<0.99:
    total =total+ pca_oliv.explained_variance_ratio_[k]
    k = k+1
k
```

```python
pca_lfw = PCA(n_components=k, whiten = True)
x_transfromd = pca_lfw.fit_transform(x_train)
x_transfromd .shape
```

```python
x_approx = pca_lfw.inverse_transform(x_transfromd)
x_approx.shape
```

```python
x_aprrox_images = x_approx.reshape((855,50,37))
```

```python
fig = plt.figure(figsize=(8,8))
for i in range(64):  
    ax = fig.add_subplot(8, 8, i + 1)
    ax.imshow(x_aprrox_images[i], cmap=plt.cm.bone)
plt.show()
```

```python
pca_lfw.components_.shape
eigenvec = pca_lfw.components_.reshape(k, 50, 37)
```

```python
fig = plt.figure(figsize=(8,8))
for i in range(64):  
    ax = fig.add_subplot(8, 8, i + 1)
    ax.imshow(eigenvec[i], cmap=plt.cm.bone)
plt.show()
```

```python
x_train_pca = x_transfromd
x_test_pca = pca_lfw.transform(x_test)
```

```python
rf = ensemble.RandomForestClassifier()
start = time.time()
rf.fit(x_train, y_train)
end = time.time()
print(end -start)
y_pred = rf.predict(x_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
rf = ensemble.RandomForestClassifier()
start = time.time()
rf.fit(x_train_pca, y_train)
end = time.time()
print(end -start)
y_pred = rf.predict(x_test_pca)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
